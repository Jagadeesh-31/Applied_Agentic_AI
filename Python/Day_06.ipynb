```python
# Lists - using methods & Functions

skills = ["Python", "Git"]

print(skills)

# Add Elements
skills.append("AI")
skills.extend(["Machine Learning", "LLMs"])
skills.insert(1, "GitHub")

print(skills)

# Access Elements
print(skills[0])
print(skills[-1])

# Slicing
print(skills[1:4])

# Update
skills[0] = "Python Programming"

print(skills)

# Remove Elements
skills.remove("Git")
skills.pop()

print(skills)

# Reverse
skills.reverse()
print(skills)

# Copy
new_skills = skills.copy()

print(new_skills)
```
