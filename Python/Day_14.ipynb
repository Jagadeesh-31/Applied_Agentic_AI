```python
# Type Conversion --> list, tuple, set, dict
# list --> str, tuple, dict

age = [23, 21, 43]
age = list((21, 12, 43))
print(type(age))
b = str(age)
print(b)
print(len(b))
c = tuple(age)
print(type(c))
e = dict.fromkeys(age)
print(e)
```

```python
# Find the sum and avg of marks

marks = list(map(int, input("Enter marks (comma separated): ").split(',')))
print(marks)
s = 0
avg = 0
for i in marks:
    print(i)
    s += i
    print(s)
avg = s / len(marks)
print(f"Total marks is {s}")
print(f"Total avg is {avg}")
```

```python
# sum of numbers entered as a comma separated string

l = list(input("Enter numbers (comma separated): ").split(','))
res = 0
for i in l:
    res += float(i)
    print(res)
print(f"Total is {res}")

details = {
    'name': ['sai', 'abhi', 'ram'],
    'marks': [45, 65, 34]
}

print(details.items())
for i in details:
    print(1)

for key in details:
    print(key)

for values in details.values():
    print(values)

for key, value in details.items():
    print(f'Keys is {key}')
    print(f'Values is {value}')
```

```python
# range (start, end, step)
for i in range(5):
    print(i, end=' ')
print()

for i in range(1, 11):
    print(i, end=' ')
print()

for i in range(10, -1, -2):
    print(i,end=' ')
print()
for i in reversed(range(0, 11, 2)):
    print(i,end=' ')
```

```python
# output: A B C D E F G H
start = input("Start letter: ").upper()
end = input("End letter: ").upper()
for i in range(ord(start), ord(end) + 1):
    print(chr(i), end=" ")
print()
print("\n")
# h f db (reverse letter range using for loop)
a = input("Start: ").lower()
b = input("End: ").lower()
for i in range(ord(b), ord(a) - 1, -2):
    print(chr(i), end=" ")
print()
```

```python
# Daily Workout --> Fitness Streak

work_log = [1, 1, 1, 0, 1, 1, 0]
longest_streak = 0
current_streak = 0
# for loop including if, else
for day in work_log:
    if day == 1:
        current_streak += 1
        if current_streak > longest_streak:
            longest_streak = current_streak
    else:
        current_streak = 0
print(longest_streak)
```
