```python
# 1. IF STATEMENT

money = int(input("Enter the billing value: "))
if money <= 100:
    print("Now you are eligible to get your items")
    print("Check again")

students = ['ram', 'akash', 'abhi', 'mani']
name = input("Enter the Student name: ").lower()
if name in students:
    marks = 50
    grade = 'A'
    print(f'{name} has secured {marks} marks and {grade} Grade')


```

```python
# 2. IF-ELSE STATEMENT

# Syntax:
# if condition:
#     statement(s)
# else:
#     statement(s)

# Example: Voter eligibility check
age = int(input("Enter the age: "))
if age >= 18:
    print("You are eligible to vote, so use it properly")
    print(f"Your age is {age} years, eligible")
else:
    print("You are not eligible to vote")
    print(f"You need to wait {18 - age} more years to get voting rights")
```

```python
# 3. IF-ELIF-ELSE STATEMENT

age = int(input("Enter the age: "))
if age >= 18:
    print("You are eligible to vote, so use it properly")
    print(f"Your age is {age} years, eligible")
elif age <= 0:
    print("Enter only positive values")
else:
    print("You are not eligible to vote")
    print(f"You need to wait {18 - age} more years to get voting rights")
```
