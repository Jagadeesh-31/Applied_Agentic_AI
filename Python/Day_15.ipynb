Problem Statement: Smart ATM and Weekend Planner System
Scenario

A bank wants to develop a Smart ATM System that provides secure access to customers and helps them plan their weekend expenses after a successful cash withdrawal.

For security reasons, every customer must verify their identity by entering the correct 4-digit ATM PIN. If the entered PIN is not exactly 4 digits, the system should show an error and re-prompt without counting it as a used attempt. The customer is allowed a maximum of three genuine attempts to enter the correct PIN. After each incorrect (but valid-format) attempt, the system must display the number of remaining attempts.

If all three attempts are used without entering the correct PIN, the customer's account is blocked for 24 hours. Since a program can't literally pause for a day, this must be simulated using a timestamp (e.g., Python's datetime module): the system records the moment the block started, and if PIN verification is attempted again before 24 hours have passed, it shows a "blocked" message along with the time remaining, instead of asking for a PIN.

If the customer enters the correct PIN within the allowed attempts, the system asks them to enter the amount they wish to withdraw. Before completing the transaction:

The withdrawal amount must be a positive number.
The account must maintain a minimum balance of ₹500 after withdrawal — i.e., the withdrawal is approved only if balance - amount ≥ 500.
If approved, deduct the amount and display both the withdrawn amount and the updated balance.
If the balance is insufficient to maintain the minimum, display an appropriate error message and do not deduct anything.

After a successful withdrawal, the ATM also acts as a Weekend Budget Planner, using these fixed ranges:

Budget Range (₹)	Suggested Activity
> 10,000	Plan a Trip
5,001 – 10,000	Resort Stay
3,001 – 5,000	Movie and Dinner
1,001 – 3,000	Cafe and Shopping
501 – 1,000	Street Food and Park Visit
0 – 500	Stay Home
Negative	Show validation message: "Please enter a valid, non-negative budget."

All numeric inputs (PIN, withdrawal amount, budget) must be validated: if the user enters non-numeric text, the system should show an error and re-prompt rather than crashing.

After the weekend activity is displayed, the program ends.

Task

Write a Python program that fulfills the above scenario using:

Variables
User input
Nested if statements
if-elif-else statements
while loop
Counter for PIN attempts

The program should:

Validate that the entered PIN is exactly 4 digits before counting it as an attempt.
Verify the ATM PIN with a maximum of 3 genuine attempts.
Display remaining attempts after each incorrect PIN.
Simulate a 24-hour account block (via timestamp) if all three attempts fail, and honor that block on any subsequent PIN check within the window.
Allow withdrawal only after successful PIN verification.
Validate the withdrawal amount (positive number, maintains ₹500 minimum balance).
Display the withdrawn amount and updated balance after a successful transaction.
Suggest a weekend activity based on the fixed budget ranges above.
Handle all invalid/non-numeric inputs gracefully, without crashing.

```python
CORRECT_PIN = 1888
balance = 15000
max_attempts = 3
attempts = 0
is_blocked = False
pin_verified = False

print("      WELCOME TO THE SMART ATM SYSTEM    ")

if is_blocked:
    print("ACCOUNT BLOCKED Your account is currently blocked for 24 hours due to 3 failed attempts.")
    print("Simulated Time Remaining: 24 hours, 00 minutes, 00 seconds.")
else:
    print("  PIN Verification ")
    while attempts < max_attempts:
        user_pin = int(input("Please enter your 4-digit ATM PIN: "))

        if user_pin < 1000 or user_pin > 9999:
            print("Error: Invalid PIN format! Must be exactly 4 digits. (Attempt not counted)\n")
            continue  

        if user_pin == CORRECT_PIN:
            print("\nSuccess: PIN verified successfully!")
            pin_verified = True
            break
        else:
            attempts += 1
            print("Error: Incorrect PIN.")

            if max_attempts - attempts > 0:
                print("Remaining attempts:", max_attempts - attempts, "\n")
            else:
                is_blocked = True
                print("\n ALERT You have exceeded 3 failed attempts. Account blocked for 24 hours!")
                print("Block Start Simulated: 24 hours, 00 minutes remaining before next login.")

if pin_verified:
    print("\n--- Cash Withdrawal ---")
    while True:
        withdrawal_amount = int(input("Enter amount to withdraw (₹): "))

        if withdrawal_amount <= 0:
            print("Error: Withdrawal amount must be a positive number greater than 0.\n")
        elif balance - withdrawal_amount < 500:
            print("Error: Transaction Declined! Account must maintain a minimum balance of ₹500.")
            print("Current Balance:", balance, "| Maximum Withdrawable Amount:", balance - 500, "\n")
        else:
            balance = balance - withdrawal_amount
            print("\n--- Transaction Receipt ---")
            print("Withdrawn Amount:", withdrawal_amount)
            print("Updated Balance :", balance)
            break

    print("--- Weekend Budget Planner ---")
    while True:
        budget = int(input("Enter your planned weekend budget (₹): "))

        if budget < 0:
            print("Please enter a valid, non-negative budget.\n")
        elif budget > 10000:
            print("\nSuggested Activity: Plan a Trip")
            break
        elif budget >= 5001:
            print("\nSuggested Activity: Resort Stay")
            break
        elif budget >= 3001:
            print("\nSuggested Activity: Movie and Dinner")
            break
        elif budget >= 1001:
            print("\nSuggested Activity: Cafe and Shopping")
            break
        elif budget >= 501:
            print("\nSuggested Activity: Street Food and Park Visit")
            break
        else:  # 0 to 500
            print("\nSuggested Activity: Stay Home")
            break

print("\nThank you for using the Smart ATM System. Program ended.")
```
