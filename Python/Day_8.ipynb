```python
"""
PYTHON: TUPLES, SETS & FROZENSETS
"""
data = (1, 24, 5)
print(data)
print(type(data))

# Nested tuple
details = ('codegnan', 32, (2, 4, 5), 'saketh', [12, 45, 'agents', 'rag'])
print(details)

print(len(details))     # Length
print(details[2])       # Indexing
print(details[4][2])    # Nested indexing

# Immutable -> can't do details[0] = ...
updated_details = (details[0].replace('n', 'f'),) + details[1:]
print(updated_details)

# List inside tuple -> mutable
details[4][2] = details[4][2].replace('a', 'A')
print(details)

print(details[1:4])   # Slicing
print(details[::3])   # Striding

print(type(details[4]))   # Type check
details[4].remove('Agents')   # Remove
print(details)
```

```python
# TUPLE OPERATIONS
age = (22, 21, 32, 25)
ids = (231, 342, 213)

print(age + ids)     # Concatenation
print(age * 2)        # Repetition
print(22 in age)      # Membership

# Built-ins
age = (25, 12, 45, 65)
print(min(age))              # Min
print(max(age))               # Max
print(tuple(sorted(age)))     # Sorted

# Methods
details = ('saketh', 'codegnan', 'Agentic AI', 34, 23, 5.8)
print(details)
print(details.index(34))   # Index
print(details.count(34))   # Count

# Typecasting
details_list = list(details)   # Tuple -> List
print(details_list)
print(type(details_list))

text = 'AgenticAI'
print(type(text))
char_list = list(text)          # String -> List
print(len(char_list))
print(char_list)
char_tuple = tuple(char_list)   # List -> Tuple
print(char_tuple)
```

```python
# 2. SETS

empty_dict = {}       # Dictionary
empty_set = set()      # Set
print(type(empty_dict))
print(type(empty_set))

ids = {123, 124, 125, 126, 127, 123, 124}
print(ids)         # Duplicates removed
print(len(ids))

# Adding
ids.add(156)              # Add
print(ids)
ids.add('agents')
print(ids)

ids.update(['saketh', 'codegnan', 124, 256])   # Update
print(ids)

names = ['Siva', 'Sam', 'Ram', 'Ajay']
ids.update(names)
print(ids)

# Removing
ids.discard(123)   # Discard
print(ids)

ids.remove(124)     # Remove
print(ids)

print(ids.pop())    # Pop

print(ids.clear())  # Clear
print(ids)
```

```python
# SET OPERATIONS

ids = {123, 124, 125, 126, 127}
ages = {35, 23, 123, 24, 45}

union_result = ids.union(ages)                       # Union
print(union_result)

intersection_result = ids.intersection(ages)          # Intersection
print(intersection_result)

difference_result = ids.difference(ages)               # Difference
print(difference_result)

symmetric_diff_result = ids.symmetric_difference(ages)  # Symmetric diff
print(symmetric_diff_result)

# Operators
print(ids | ages)   # Union
print(ids & ages)    # Intersection
print(ids - ages)    # Difference
print(ids ^ ages)    # Symmetric diff

# Boolean checks
a = {1, 2, 3}
b = {1, 2, 3, 4, 5}
print(a.issubset(b))     # Subset
print(b.issuperset(a))    # Superset
print(a.isdisjoint(b))    # Disjoint
```

```python
# 3. FROZENSET

data = frozenset(ids)
print(data)
print(type(data))

temp_details = frozenset([34, 35, 34, 32, 31])
print(temp_details)
print(min(temp_details))   # Min
print(max(temp_details))    # Max
print(sorted(temp_details))   # Sorted
```
