```python
# 1. DICTIONARY BASICS

details = {}
print(details)
print(type(details))

details = {'Name': 'Codegnan', 'Place': 'Hyd', 'age': 7}
print(details)
print(len(details))

print(details['Name'])   # Access
# print(details['Age'])   # KeyError

# Unique keys
data = {'Age': 25, 'name': 'code', 'Age': 26}
print(data)   # Latest value kept
```

```python
# 2. DICTIONARY WITH OTHER DATATYPES

students_data = {
    'ids': [23, 21, 45, 52],
    'names': ['Pream', 'Saki', 'Varada', 'Sushas'],
    'place': ('Hyd', 'Vjwda'),
    'gender': {'male', 'female'}
}

print(len(students_data))
print(students_data.keys())      # Keys
print(students_data['names'])
print(students_data.values())     # Values

students_data['Course'] = ['PFS', 'JFS', 'AAA', 'DA']   # Add
print(students_data)
print(type(students_data))
print(type(students_data['ids']))
```

```python
# 3. MODIFYING NESTED VALUES

students_data['ids'].extend([56, 67, 87])          # Extend
print(students_data['ids'])

students_data['names'].insert(1, 'Ashok')           # Insert
print(students_data['names'])

students_data['place'] = list(students_data['place'])   # Tuple -> List
print(students_data['place'])

students_data['place'].append('Vizag')             # Append
print(students_data['place'])

students_data['names'].sort()                        # Sort
print(students_data['names'])

print(students_data['Course'][1::2])   # Slicing

del students_data['ids'][1::3]   # Delete
print(students_data['ids'])
```

```python
# 4. keys(), values(), items()
print(students_data.items())   # Items

print(students_data.get('branch'))           # Get
print(students_data.get('branch', 'CSE'))     # Default
print(students_data.get('names'))

print(students_data.setdefault('ids'))                    # Setdefault
students_data.setdefault('branch', ['CSE', 'CSD', 'ECE', 'IT'])
print(students_data)
```

```python
# 5. update(), pop(), popitem(), clear(), copy()

students_data.update({'fees': [456, 234], 'marks': [45, 78, 85]})   # Update
print(students_data)

print(students_data.pop('marks'))    # Pop
print(students_data)

print(students_data.popitem())        # Popitem
print(students_data)

backup = students_data.copy()   # Copy
print(backup)

backup.clear()   # Clear
print(backup)

```

```python
# 6. fromkeys()

ids = [23, 45, 67]
d = dict.fromkeys(ids)   # Fromkeys
print(d)

d[23] = 'random'
print(d)

print(23 in d)   # Membership
```

```python
# NESTED DICTIONARIES

data = {
    's1': {'id': 23, 'name': 'ram', 'place': 'hyd'},
    's2': {'id': 25, 'name': 'sony', 'place': 'bng'}
}

employees = {
    'e1': {'id': 101, 'role': 'Data Analyst', 'salary': 45000},
    'e2': {'id': 102, 'role': 'ML Engineer', 'salary': 65000}
}

print("Original Data:", data)
print()

# 1. keys()
print("Keys:", data.keys())

# 2. values()
print("\nValues:")
print(data.values())

# 3. items()
print("\nItems:")
print(data.items())

# 4. get()
print("\nUsing get():")
print(data.get('s1'))
print(data.get('s3'))                  # None
print(data.get('s3', 'Not Found'))

# 5. update()
print("\nUsing update():")
data.update({
    's3': {'id': 30, 'name': 'krishna', 'place': 'vizag'}
})
print(data)

# 6. pop()
print("\nUsing pop():")
removed = data.pop('s3')
print("Removed:", removed)
print(data)

# 7. popitem()
temp = data.copy()
print("\nUsing popitem():")
print(temp.popitem())
print(temp)

# 8. copy()
print("\nUsing copy():")
copy_data = data.copy()
print(copy_data)

# 9. clear()
temp = data.copy()
temp.clear()
print("\nUsing clear():")
print(temp)

# 10. setdefault()
print("\nUsing setdefault():")
data.setdefault('s4', {'id': 40, 'name': 'Raju', 'place': 'Delhi'})
print(data)

# 11. fromkeys()
print("\nUsing fromkeys():")
students = dict.fromkeys(['s5', 's6'], {'id': 0, 'name': '', 'place': ''})
print(students)

# 12. len()
# 
print("\nLength:", len(data))

# 13. type()
print("\nType:", type(data))

# 14. Access Nested Dictionary
print("\nNested Access:")
print(data['s1']['name'])
print(data['s2']['place'])

# 15. Modify Nested Value
print("\nModify Nested Value:")
data['s1']['place'] = "Hyderabad"
print(data['s1'])

# 16. Add New Key Inside Nested Dictionary
print("\nAdd New Nested Key:")
data['s1']['course'] = "Python"
print(data['s1'])

# 17. Delete Nested Key
print("\nDelete Nested Key:")
del data['s1']['course']
print(data['s1'])

# 18. Membership Operators
print("\nMembership:")
print('s1' in data)
print('s5' in data)

 19. Loop Through Outer Dictionary
# ===================================
print("\nOuter Dictionary Loop:")
for key, value in data.items():
    print(key, value)

# ===================================
# 20. Loop Through Nested Dictionary
# ===================================
print("\nNested Dictionary Loop:")
for student, details in data.items():
    print(student)
    for k, v in details.items():
        print(k, ":", v)

# ===================================
# 21. Employees Dictionary
# ===================================
print("\nEmployees:")
print(employees)

print("\nEmployee Keys:")
print(employees.keys())

print("\nEmployee Values:")
print(employees.values())

print("\nEmployee Items:")
print(employees.items())

print("\nAccess Nested Value:")
print(employees['e2']['role'])

print("\nUpdate Salary:")
employees['e1']['salary'] = 50000
print(employees)

print("\nAdd Department:")
employees['e2']['department'] = "AI"
print(employees)

print("\nDelete Department:")
del employees['e2']['department']
print(employees)

print("\nLoop Employees:")
for emp, details in employees.items():
    print(emp)
    for k, v in details.items():
        print(k, ":", v)

# ===================================
# 22. Dictionary Comprehension
# ===================================
print("\nDictionary Comprehension:")
salary = {emp: details['salary'] for emp, details in employees.items()}
print(salary)

# ===================================
# 23. Maximum Salary
# ===================================
print("\nMaximum Salary:")
print(max(salary.values()))

# ===================================
# 24. Minimum Salary
# ===================================
print("\nMinimum Salary:")
print(min(salary.values()))

# ===================================
# 25. Sum of Salaries
# ===================================
print("\nTotal Salary:")
print(sum(salary.values()))

# ===================================
# 26. Sorted Keys
# ===================================
print("\nSorted Keys:")
print(sorted(data.keys()))

# ===================================
# 27. Sorted Employees
# ===================================
print("\nSorted Employee Keys:")
print(sorted(employees.keys()))
```
