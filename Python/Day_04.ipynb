```python
## Input & Output Formatting

# User Input

name = input("Enter your Name: ")

print("\n----- User Details -----")
print("Name :", name)

# Multiple Inputs
city, course = input("\nEnter City and Course: ").split(",")

print("\nCity :", city)
print("Course :", course)

# Multiple Integer Inputs
a, b = map(int, input("\nEnter two numbers: ").split(","))

print("Addition :", a + b)
print("Subtraction :", a - b)

# Formatted Output
print(f"\n{name} is learning {course} at {city}.")
```
