```python
# 1. LISTS

fruits = ['apple', 'banana', 'cherry']
print(fruits)
print(type(fruits))

print(fruits[0])       # Indexing
print(fruits[-1])      # Negative indexing
print(fruits[0:2])     # Slicing

# Mutable -> can reassign
fruits[1] = 'blueberry'
print(fruits)

# Adding
fruits.append('mango')            # Append
print(fruits)

fruits.insert(1, 'kiwi')          # Insert
print(fruits)

fruits.extend(['grape', 'fig'])   # Extend
print(fruits)

# Removing
fruits.remove('kiwi')             # Remove
print(fruits)

popped_item = fruits.pop()        # Pop
print(popped_item)
print(fruits)

# Other operations
print(len(fruits))                 # Length
print('mango' in fruits)           # Membership
fruits.sort()                       # Sort
print(fruits)
fruits.reverse()                    # Reverse
print(fruits)

```

```python
# 2. PRACTICE: NESTED SEQUENCE

nested_sequence = [
    'Python',                          # String
    ('Tuples', 'are', 'immutable'),    # Tuple
    {'Sets', 'are', 'unique'},         # Set
    'Lists are mutable'                # String
]

print(nested_sequence)

# Loop & check types
for item in nested_sequence:
    print(item, '->', type(item))

print(nested_sequence[1][0])           # Nested indexing
print('Sets' in nested_sequence[2])     # Membership

# Mutable -> can modify
nested_sequence.append('New item added to the list')
print(nested_sequence)
```
