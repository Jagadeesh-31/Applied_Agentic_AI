```python
# Operators

a = 20
b = 6

print("Arithmetic")
print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a // b)
print(a % b)
print(a ** 2)

print("Comparison")
print(a > b)
print(a == b)

print("Logical")
print(a > 10 and b < 10)

print("Membership")
numbers = [10, 20, 30]
print(20 in numbers)

print("Identity")
x = numbers
print(x is numbers)
```
