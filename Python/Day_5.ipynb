```python
# Strings - methods & Functions

text = "Agentic AI"

print("Original :", text)

# Case Conversion
print(text.upper())
print(text.lower())
print(text.title())
print(text.capitalize())
print(text.swapcase())

# Indexing
print(text[0])
print(text[-1])

# Slicing
print(text[:8])
print(text[9:])
print(text[::-1])

# Searching
print(text.find("AI"))
print(text.count("A"))

# Replace
print(text.replace("AI", "Python"))

# Split
sentence = "Python,AI,ML,LLM"
print(sentence.split(","))

# Join
words = ["Learn", "Python", "Daily"]
print(" ".join(words))

# Strip
name = "   Jagadeesh   "
print(name.strip())
```
