```python
"""
DAY 8 — PYTHON: TUPLES, SETS & FROZENSETS
------------------------------------------
A quick recap of the core concepts practiced today, with clean,
runnable examples and explanations for each.
"""

# ============================================================
# 1. TUPLES
# ============================================================
# A tuple is an IMMUTABLE, ORDERED, INDEXED, HETEROGENEOUS
# sequence type. Once created, its elements cannot be changed.
# Declared using () — parentheses are optional but recommended.

data = (1, 24, 5)
print(data)        # (1, 24, 5)
print(type(data))  # <class 'tuple'>

# --- Nested tuples (can hold tuples, lists, or any other type) ---
details = ('codegnan', 32, (2, 4, 5), 'saketh', [12, 45, 'agents', 'rag'])
print(details)

print(len(details))     # 5 -> total number of elements
print(details[2])       # (2, 4, 5) -> nested tuple returned as-is
print(details[4][2])    # 'agents' -> indexing into the nested list

# Tuples are immutable, so this is NOT allowed:
# details[0] = details[0].replace('n', 'f')   # ❌ TypeError
# Instead, create a NEW tuple with the modified value:
updated_details = (details[0].replace('n', 'f'),) + details[1:]
print(updated_details)

# Lists INSIDE a tuple are still mutable (only the tuple structure is fixed)
details[4][2] = details[4][2].replace('a', 'A')
print(details)  # notice index 4's list has changed

# --- Slicing & striding ---
print(details[1:4])   # elements from index 1 to 3
print(details[::3])   # every 3rd element

print(type(details[4]))   # always check type before operating on nested items
details[4].remove('Agents')   # note: it's 'Agents' now, since we capitalized it above
print(details)

# ============================================================
# TUPLE OPERATIONS: Indexing, Slicing, Membership,
#                    Concatenation, Repetition
# ============================================================
age = (22, 21, 32, 25)
ids = (231, 342, 213)

print(age + ids)     # concatenation -> merges two tuples
print(age * 2)        # repetition -> repeats the tuple
print(22 in age)      # membership test -> True

# --- Built-in functions: len(), type(), min(), max(), sorted() ---
age = (25, 12, 45, 65)
print(min(age))              # smallest element
print(max(age))               # largest element
print(tuple(sorted(age)))     # sorted() returns a list, so we typecast back to tuple

# --- Tuple methods: index(), count() ---
details = ('saketh', 'codegnan', 'Agentic AI', 34, 23, 5.8)
print(details)
print(details.index(34))   # position of first occurrence of 34
print(details.count(34))   # number of times 34 appears

# --- Typecasting: Tuple <-> List <-> String ---
details_list = list(details)
print(details_list)
print(type(details_list))

text = 'AgenticAI'
print(type(text))
char_list = list(text)
print(len(char_list))
print(char_list)
char_tuple = tuple(char_list)
print(char_tuple)


# ============================================================
# 2. SETS
# ============================================================
# A set is a MUTABLE, UNORDERED collection of UNIQUE elements.
# Declared using set() — NOT {} (that creates an empty dictionary).

empty_dict = {}       # this is a dictionary by default
empty_set = set()      # this is how you create an empty set
print(type(empty_dict))
print(type(empty_set))

ids = {123, 124, 125, 126, 127, 123, 124}
print(ids)         # duplicates are automatically removed
print(len(ids))

# Note: a set cannot contain unhashable types like lists or other sets
# data = {23, 4.5, 'codegnan', {12, 34, 5}}   # ❌ TypeError

# --- Adding elements: add(), update() ---
ids.add(156)
print(ids)
ids.add('agents')
print(ids)

ids.update(['saketh', 'codegnan', 124, 256])   # add multiple elements at once
print(ids)

names = ['Siva', 'Sam', 'Ram', 'Ajay']
ids.update(names)
print(ids)

# --- Removing elements: discard(), remove(), pop(), clear() ---
ids.discard(123)   # removes if present, no error if missing
print(ids)

ids.remove(124)     # removes, but raises KeyError if not present
print(ids)
# ids.remove(999)   # ❌ KeyError -> use discard() to avoid this risk

print(ids.pop())    # removes & returns an arbitrary element (sets are unordered)

print(ids.clear())  # removes everything, returns None
print(ids)           # now an empty set

# ============================================================
# SET OPERATIONS: Union, Intersection, Difference,
#                  Symmetric Difference, Subset, Superset
# ============================================================
ids = {123, 124, 125, 126, 127}
ages = {35, 23, 123, 24, 45}

union_result = ids.union(ages)                       # all unique elements from both
print(union_result)

intersection_result = ids.intersection(ages)          # common elements only
print(intersection_result)

difference_result = ids.difference(ages)               # in ids but not in ages
print(difference_result)

symmetric_diff_result = ids.symmetric_difference(ages)  # in either, but not both
print(symmetric_diff_result)

# --- Shorthand operators: | (union), & (intersection),
#     - (difference), ^ (symmetric difference) ---
print(ids | ages)
print(ids & ages)
print(ids - ages)
print(ids ^ ages)

# --- Boolean relationship checks ---
a = {1, 2, 3}
b = {1, 2, 3, 4, 5}
print(a.issubset(b))     # True -> every element of a is in b
print(b.issuperset(a))    # True -> b contains all elements of a
print(a.isdisjoint(b))    # False -> they share common elements


# ============================================================
# 3. FROZENSET — an IMMUTABLE version of a set
# ============================================================
# Supports read-only operations (math operations, membership),
# but you cannot add or remove elements after creation.

data = frozenset(ids)
print(data)
print(type(data))

temp_details = frozenset([34, 35, 34, 32, 31])
print(temp_details)
print(min(temp_details))
print(max(temp_details))
print(sorted(temp_details))   # returns a sorted list
```
