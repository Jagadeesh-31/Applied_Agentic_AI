```python
# 1. Assert Statement

x = int(input("Enter a Positive Number: "))

# Assertion
assert x > 0, "Value must be a Positive Number"

x += 2

print("Updated Value:", x)
```

```python
# 2. Nested Loop Example

for i in range(3):
    for j in range(2):
        print(i, j)
```

```python
# 3. Number Pattern

# 1 2 3
# 1 2 3
# 1 2 3

for i in range(1, 5):
    for j in range(1, 4):
        print(j, end=" ")
    print()
```

```python
# 4. Row Number Pattern
# 1 1 1
# 2 2 2
# 3 3 3

for i in range(1, 4):
    for j in range(1, 4):
        print(i, end=" ")
    print()
```

```python
 # 5. Alphabet Pattern
# A A A
# B B B
# C C C

for i in range(ord('A'), ord('D')):
    for j in range(ord('A'), ord('D')):
        print(chr(i), end=" ")
    print()
```

```python
# 6. Sequential Number Pattern
# 1 2 3
# 4 5 6
# 7 8 9

num = 1

for i in range(3):
    for j in range(3):
        print(num, end=" ")
        num += 1
    print()
```

```python
# 7. Square Star Pattern (3 × 3)
# * * *
# * * *
# * * *

for i in range(3):
    for j in range(3):
        print("*", end=" ")
    print()
```

```python
# 8. Rectangle Star Pattern (4 × 5)


for i in range(4):
    for j in range(5):
        print("*", end=" ")
    print()
```

```python
# 9. Right Triangle Star Pattern
#
# *
# **
# ***
# ****

for i in range(1, 5):
    for j in range(i):
        print("*", end=" ")
    print()
```
